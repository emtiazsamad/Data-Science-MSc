```python
import numpy as np
import pandas as pd
import datetime
from datetime import datetime as dt
```

```python
df = pd.read_excel('completed_form.xlsm')
df = df.fillna(value='')
df.columns = df.iloc[0]
df = df[1:]
```

```python
complete_data = pd.DataFrame(columns=['PrimaryActivityCode', 'SecondaryActivityCode', 'DescriptiveLabel', 'StartDate', 'UniBathWeekNo', 'DayOfWeek', 'StartTime', 'EndTime', 'DurationMins', 'EnjoymentScore', 'DietRating', 'SleepQuality'])
complete_data['PrimaryActivityCode'] = df['PRIMARY ACTIVITY']
complete_data['SecondaryActivityCode'] = df['SECONDARY ACTIVITY']
complete_data['DescriptiveLabel'] = df['DESCRIPTION']
complete_data['StartDate'] = df['DATE']
complete_data['StartTime'] = df['START TIME']
complete_data['EndTime'] = df['END TIME']
complete_data['EnjoymentScore'] = df['ENJOYMENT']
complete_data['DietRating'] = df['DIET RATING']
complete_data['SleepQuality'] = df['SLEEP QUALITY']
```

```python
complete_data
```

```python
complete_data['StartDate'] = pd.to_datetime(complete_data['StartDate'])

d = ((complete_data['StartDate'].dt.date - 
      dt.strptime('2022-10-04', '%Y-%m-%d').date() + datetime.timedelta(days=365)).dt.days)

complete_data['UniBathWeekNo'] = (np.floor(d/7)+1).astype(int)
complete_data['DayOfWeek'] = complete_data['StartDate'].dt.day_name()
```

```python
complete_data
```

```python
complete_data['StartDate'] = complete_data['StartDate'].dt.strftime('%d/%m/%Y')

start = pd.to_datetime(complete_data['StartDate'] + ' ' + complete_data['StartTime'].astype(str))
end = pd.to_datetime(complete_data['StartDate'] + ' ' + complete_data['EndTime'].astype(str))
end = end.apply(lambda x: x + datetime.timedelta(days = 1) if x.time() == datetime.time(0, 0) else x)
duration = (end - start).dt.total_seconds()/60

complete_data['DurationMins'] = duration.astype(int)
complete_data['StartTime'] = start.dt.strftime('%H:%M')
complete_data['EndTime'] = end.dt.strftime('%H:%M')
```

```python
complete_data
```

```python
complete_data.to_csv('complete_data.csv', index = False)
```
